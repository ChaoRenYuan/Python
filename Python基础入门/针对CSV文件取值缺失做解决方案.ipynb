```python
import csv

path1 = "/Users/rendashu/Desktop/商铺数据.csv"

test = ",;。"
testList = []
with open(path1, newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',')
    for row in spamreader:
        testList.append(test.join(row))
# print(testList)
```

```python
# 获取评论
def getComment(data):
    s = "条"
    if s in data:
        s_res = data.split(s)[0].strip()
        return s_res
    else:
        return "数据缺失"

# 获取评论清单
def getCommentList(data):
   
    if len(data) == 3:
        quality = data[0][2:]
        environment = data[1][2:]
        service = data[-1][2:].replace("\n", "")
        return [quality, environment, service]
    else:
        return "数据缺失"
   
# 获取价格
def getPrice(data):

    if "￥" in data:
        price = data.split("￥")
        return price[-1]
    else:
        return "数据缺失"


    

data_res = []    
n = 0
for line in testList:

    n += 1
    comment = line.split(test)[2]           
    commentList = line.split(test)[-1].split("                                ")
    price = line.split(test)[4]
    
    # 获取评论
    comment_res = getComment(comment)
    # 获取评论清单
    commentList_res = getCommentList(commentList)
    # 获取价格
    price_res = getPrice(price)
    
    
    classify = line.split(test)[0]
    name = line.split(test)[1]
    
    lineList = line.split(test)
    address = lineList[5]


    if "数据缺失" not in [comment_res, commentList_res, price_res]:
        data = [["classify", classify],
               ["name", name],
               ["comment", comment_res],
               ["price", price_res],
               ["address", address],
               ["commentList", commentList_res]]
        data_res.append(dict(data))
print(data_res)

```
