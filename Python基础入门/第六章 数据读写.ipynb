```python
# 本地文件的界定：指向一个本地存储的文件，是一个链接或者一个映射

path1 = "/Users/zorn/Desktop/test.txt"
print(path1)


# PS:以下是windows系统的栗子
# path1 = 'C:/Users/zorn/Desktop/text.txt'  # 单个反斜杠：/
# path2 = 'C:\\Users\\zorn\\Desktop\\text.txt'  # 两个斜杠：\\（第一个\是转义符）
# path3 = r'C:\Users\zorn\Desktop\text.txt'  # r用于防止字符转义
```

```python
# 读取文件：open语句

f = open(path1, "r", encoding="utf8") # 这里的r是只读的意思
print(type(f))
print(f.read())
print("读取完毕")
# open("路径", "模式", enconding = "编码")
# 模式：r:读取文件，默认；w: 写入文件; rw: 读取+写入； a：追加
# 简答的读取方法：.read() -> 读取后，光标将会留在读取末尾

print(f.read())
print("读取为空")
# 运行第一次.read()之后，光标位于末尾，再次读取输出为空

f.seek(0)
print(f.read())
print("第二次读取")
# 所以现在用f.seek(0) 来移动光标

f.close()
# print(f.read()) # 关闭后无法读取
# 关闭文件链接 f.close(), 养成一个好习惯
```

```python
'''
小作业
利用百度POI小插件爬取一些poi数据，然后存成txt，再用python读取，编写成一个json形式（列表字典）：
[{'name':'...', 'lng':..., 'lat':..., 'address':'...'},{...},...,]
'''
```

```python
# 答案代码

path = '/Users/zorn/Desktop/jiuba.txt'  # 创建路径变量
f = open(path, 'r')  # 读取文件
m = [] # 新建一个空列表，用于存储数据
n = 0

for line in f.readlines():

    n += 1
    st1 = line.split(":")
    name = st1[0] # 先以:拆分，筛选出name文本
    information = st1[1] # 第二部分包括lng, lat, address
    
    st2 = information.split(",")
    lng = float(st2[0]) # 筛选出lng
    lat = float(st2[1]) # 筛选出lat
    ad = st2[2].strip() # 筛选出address,去除末尾空格
    
    result =[["name", name], ["lng", lng], ["lat", lat], ["address", ad]] # 做成嵌套列表
    m.append(dict(result)) # 生成字典，添加到m
print(m)
    
```

```python
# 文件写入 - write

path = "/Users/zorn/Desktop/aaa.txt"
f = open(path, "w", encoding = "utf8")
f.write("我爱中国🇨🇳")
# 执行后直接创建"aaa.txt"文件，并写入“我爱中国🇨🇳”
# 6代表输入了6个字符，可以打开看看     🇨🇳(两字符)
```

```python
path = "/Users/zorn/Desktop/"
f = open(path + "aaa.txt", "w", encoding = "utf8")
lst = ["a", "b", "c", "d", "e"]
f.writelines(lst)
f.close() 
# f.writelines(list): 依次把列表元素写入，但这里没有换行？
# 如果想要每一个元素换行怎么办？-> 每个元素后都添加"\n"
```

```python
path = "/Users/zorn/Desktop/"
f = open(path + "aaa.txt", "w", encoding = "utf8")
lst = ["a", "b", "c", "d", "e"]

for i in range(len(lst)):
    lst[i] = lst[i] + "\n"
f.writelines(lst)
f.close()
# 通过遍历给每个元素最后加上“\n”
```

```python
'''
小作业
两个列表[1~10],[a~j]，写入一个txt，变成以下格式
1,a
2,b
3,c
...
'''
```

```python
# 答
aList = list(range(1,11))
bList = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j"]
cList = []
path = "/Users/Zorn/Desktop/"
f = open(path + "bbb.txt", "w", encoding="utf8")
for i in range(len(aList)):
    cList.append("{},{}\n".format(aList[i], bList[i]))
f.writelines(cList)
f.close()
```

```python
'''
【课程6.4】  pickle模块的运用

pickle提供了一个简单的持久化功能。可以将对象以文件的形式存放在磁盘上。

python的pickle模块实现了基本的数据序列和反序列化

通过pickle模块的序列化操作我们能够将程序中运行的对象信息保存到文件中去，永久存储

通过pickle模块的反序列化操作，我们能够从文件中创建上一次程序保存的对象。

pickle.dump() / pickle.load()

'''
```

```python
# 存储：pickle.dump(obj, file, [, protocol])

import pickle

data = {"a":[1,2,3,4], "b":("string","abc"), "c":"hello"}
print(data)
# 创建一个字典变量data

pic = open("/Users/zorn/Desktop/data.pkl", "wb")
# 以二进制来存储：rb, wb, wrb, ab

pickle.dump(data, pic)
pic.close()
# 将一个字典数据存成了pkl文件（名字可自己取）
```

```python
# 读取：pickle.load(file)

f = open("/Users/zorn/Desktop/data.pkl", "rb")

st = pickle.load(f)
print(st)
# 直接读取pkl文件内的数据，该数据为一个字典
```
