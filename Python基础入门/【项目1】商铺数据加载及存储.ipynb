```python
'''
【项目1】  商铺数据加载及存储

作业要求：
1、成功读取“商铺数据.csv”文件
2、数据清洗：
① comment，price两个字段清洗成数字
② commentlist拆分成三个字段，并且清洗成数字
③ 清除字段缺失的数据
3、解析数据，存成列表字典格式：[{'var1':value1,'var2':value2,'var3':values,...},...,{}]
4、结果存为.pkl文件

'''
```

```python
path = "/Users/rendashu/Desktop/商铺数据.csv"
file = open(path, "r", encoding="utf-8")
data_res = []
# print(file.read()[:2])
for line in file.readlines()[0:2]:
    print(line)
```

```python
# 数据清洗

# 获取评论
def getComment(data):
    s = "条"
    if s in data:
        s_res = data.split(s)[0].strip()
        return s_res
    else:
        return "数据缺失"

# 获取评论清单
def getCommentList(data):
   
    if len(data) == 3:
        quality = data[0][2:]
        environment = data[1][2:]
        service = data[-1][2:].replace("\n", "")
        return [quality, environment, service]
    else:
        return "数据缺失"
   
# 获取价格
def getPrice(data):

    if "￥" in data:
        price = data.split("￥")
        return price[-1]
    else:
        return "数据缺失"
    
file.seek(0)    
for line in file.readlines()[1:]:
    
    comment = line.split(",")[2]
    commentList = line.split(",")[-1].split("                                ")
    price = line.split(",")[4]
    
    # 获取评论
    comment_res = getComment(comment)
    # 获取评论清单
    commentList_res = getCommentList(commentList)
    # 获取价格
    price_res = getPrice(price)
    
    classify = line.split(",")[0]
    name = line.split(",")[1]
    address = line.split(",")[5]
    
    if "数据缺失" not in [comment_res, commentList_res, price_res]:
        data = [["classify", classify],
               ["name", name],
               ["comment", comment_res],
               ["price", price_res],
               ["address", address],
               ["commentList", commentList_res]]
        data_res.append(dict(data))
print(data_res)
```

```python
# 存储数据
import pickle

pic = open("/Users/rendashu/Desktop/data.pkl", "wb")
pickle.dump(data_res, pic)
pic.close()
print("完成")

pic1 = open("/Users/rendashu/Desktop/data.pkl", "rb")
st = pickle.load(pic1)
print(st)
```
